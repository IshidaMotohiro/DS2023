```python
import pandas as pd
import numpy as np
file = '/home/ishida/OneDrive/2023/class/大学院/A000010_shussekibo_group.xlsx'
dat = pd.read_csv(file)
```

```python
with open(file,mode="rb") as excel_file:
    df = pd.read_excel(file)
```

```python
df.head()
```

```python
## calculate modulo 9 of index and set this as a new column
df['group'] = df.index % 9
```

```python
df.head()
```

```python
df_shuffled = df.groupby('group').apply(lambda x: x.sample(frac=1)).reset_index(drop=True)
```

```python
# divide rows into groups and randomly shuffle them and set numbers within each group and save to a new dataframe
df_shuffled['number'] = df_shuffled.groupby('group').cumcount()+1 
df_shuffled['number'] = df_shuffled['number'] % 7 + 1

```

```python
df_shuffled.head(10)
```

```python
df_shuffled.sort_values(by=['group', 'number'], inplace=True)
```

```python
df_shuffled.to_excel('A000010_shussekibo_group_shuffled.xlsx', index=False)
```

```python
grouped = df_shuffled.groupby('group')
for group in grouped.groups:
    print(group)
    # print(grouped.get_group(group), "\n")
    grouped = group.groupby('number').cumcount()+1
    grouped.get_group(group).to_csv("group_" + str(group) + ".csv", index=False)
```

```python
# divide rows into groups and save each group as a new dataframe
grouped = df.groupby('group')
for group in grouped.groups:
    print(group)
    print(grouped.get_group(group), "\n")
    grouped.get_group(group).to_csv("group_" + str(group) + ".csv", index=False)
```
